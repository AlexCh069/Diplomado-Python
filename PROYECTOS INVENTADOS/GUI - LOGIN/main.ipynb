```python
# Creacion de interfaz (GUI)

import tkinter as tk

def nueva_ventana():
    ventana = tk.Tk()
    lbl2 = tk.Label(ventana, text = "Modificado")
    lbl2.pack()
    ventana.mainloop()


ventana = tk.Tk()
ventana.title("prueba de modificacion")

lbl1 = tk.Label(ventana, text = "Prueba de modificacion")
lbl1.pack()

btn_clear = tk.Button(ventana, text = "CAMBIAR VENTANA", command= nueva_ventana)
btn_clear.pack()

ventana.mainloop()
```

```python
import tkinter as tk

def borrar_contenido():
    # Esta función borra todos los widgets dentro de la ventana
    for widget in ventana.winfo_children():
        widget.destroy()
    lbl = tk.Label(ventana, text = "BORRADO")
    lbl.pack()

# Crear la ventana
ventana = tk.Tk()
ventana.title("Ejemplo de Tkinter")

# Crear algunos widgets dentro de la ventana
label = tk.Label(ventana, text="Hola, esta es una ventana con contenido.")
label.pack()

boton_borrar = tk.Button(ventana, text="Borrar Contenido", command=borrar_contenido)
boton_borrar.pack()

# Ejecutar el bucle principal de la ventana
ventana.mainloop()
```

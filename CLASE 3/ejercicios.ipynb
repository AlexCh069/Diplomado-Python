```python
#EJEMPLO DE LABEL Y BUTTON 
import tkinter as tk

class Aplicacion:
    def __init__(self):
        self.valor=1
        self.ventana1=tk.Tk() #creo la ventana
        self.ventana1.title("Controles Button y Label")

        self.label1=tk.Label(self.ventana1,text=self.valor)
        self.label1.grid(column=0,row=0)
        self.label1.configure(foreground="red")

        self.boton1=tk.Button(self.ventana1, text="Incrementar", command=self.incrementar)
        self.boton1.grid(column=0, row=1)

        self.boton2=tk.Button(self.ventana1, text="Decrementar", command=self.decrementar)
        self.boton2.grid(column=0, row=2)

        self.ventana1.mainloop()
        

 
    def incrementar(self):
        self.valor=self.valor+1
        self.label1.config(text=self.valor)

    def decrementar(self):
        self.valor=self.valor-1
        self.label1.config(text=self.valor)      

#PROGRAMA PRINCIPAL 
aplicacion1=Aplicacion()
```

```python
#Mostrar dos controles de tipo Radiobutton con las etiquetas "Var贸n" y "Mujer", 
#cuando se presione un bot贸n actualizar una Label con el Radiobutton seleccionado.
import tkinter as tk

class Aplicacion:
    def __init__(self):
        self.ventana1=tk.Tk()
        self.ventana1.title("EJERCICIO DE RADIOBUTTON")
        
        self.seleccion=tk.IntVar()  # Instanciamos a una variable tipo int
        self.seleccion.set(1)       # inicializamos la variable con valor == 1 
        
        self.radio1=tk.Radiobutton(self.ventana1,
                                   text="Varon",    # Texto del RadioButton
                                   variable=self.seleccion, # Variable asociada al radio button 
                                   value=0)         # Valor de seleccion del radiobutton
        # NOTA: Cuando la variable de seleccion y el value coinciden se seleccionara el radiobutton 
        self.radio1.grid(column=0, row=0)
        
        self.radio2=tk.Radiobutton(self.ventana1,
                                   text="Mujer", 
                                   variable=self.seleccion, 
                                   value=1)
        self.radio2.grid(column=0, row=1)
        
        self.boton1=tk.Button(self.ventana1, text="Mostrar seleccionado", command=self.mostrarseleccionado)
        self.boton1.grid(column=0, row=2)
        
        self.label1=tk.Label(self.ventana1,text="opcion seleccionada")
        self.label1.grid(column=0, row=3)
        
        self.ventana1.mainloop()

    def mostrarseleccionado(self):
        if self.seleccion.get()==0:
            self.label1.configure(text="opcion seleccionada=Varon")
        if self.seleccion.get()==1:
            self.label1.configure(text="opcion seleccionada=Mujer")

aplicacion1=Aplicacion()
```

```python
import tkinter as tk

class app:
    def __init__(self):
        self.window = tk.Tk()
        self.window.title(" RadioButtons")

        self.lbl1 = tk.Label(self.window, text = "Modificamos RadioButtons") 
        self.lbl1.grid(row= 0, column= 0)
        # _________________________________________________________________
        self.frm_rb = tk.Frame(self.window)
        self.frm_rb.grid(row= 1, column= 0)

        self.selec=tk.IntVar()  # Instanciamos a una variable tipo int
        self.selec.set(1)       # inicializamos la variable con valor == 1 

        self.rb_1 = tk.Radiobutton(self.frm_rb, text="Option 1", variable = self.selec, value = 1,command= self.comand_rt)
        self.rb_1.grid(row= 0, column= 0)

        self.rb_2 = tk.Radiobutton(self.frm_rb, text="Option 2", variable = self.selec, value= 2, command= self.comand_rt)
        self.rb_2.grid(row= 1, column= 0)
        # ________________________________________________________________

        self.frm_packs = tk.Frame(self.window, bd = 3, relief= tk.SUNKEN)
        self.frm_packs.grid(row= 1, column = 1)

        self.lbl2 = tk.Label(self.frm_packs, text = "Modificadores")
        self.lbl2.pack()
        #________________________________________________________________

        self.btn_close = tk.Button(self.window, text = "X", command= self.destroy_window)
        self.btn_close.grid(row = 2, column= 0)


        self.window.mainloop()
    
    def comand_rt(self):
        if self.selec.get() == 1:
            self.lbl2.config(text= "Opcion 1")
        else:
            self.lbl2.config(text= "Opcion 2")
    
    def destroy_window(self):
        self.window.destroy()
    


window = app()
```

```python
#Mostrar una ventana y en su interior tres controles de tipo Checkbutton
#cuyas etiquetas correspondan a distintos lenguajes de programaci贸n. 
#Cuando se presione un bot贸n mostrar en una 
#Label la cantidad de Checkbutton que se encuentran chequeados.
import tkinter as tk
class Aplicacion:
    def __init__(self):
        self.ventana1=tk.Tk()
        self.ventana1.title("EJERCICIO DE CHECKBUTON ")

        self.seleccion1=tk.IntVar()
        self.check1=tk.Checkbutton(self.ventana1,text="Python", variable=self.seleccion1, command= self.print_name)
        self.check1.grid(column=0, row=0)

        self.seleccion2=tk.IntVar()
        self.check2=tk.Checkbutton(self.ventana1,text="C++", variable=self.seleccion2, command= self.print_name)
        self.check2.grid(column=0, row=1)

        self.seleccion3=tk.IntVar()
        self.check3=tk.Checkbutton(self.ventana1,text="Java", variable=self.seleccion3, command= self.print_name)
        self.check3.grid(column=0, row=2)

        self.boton1=tk.Button(self.ventana1, text="Verificar", command=self.verificar)
        self.boton1.grid(column=0, row=4)

        self.label1=tk.Label(self.ventana1,text="cantidad:")
        self.label1.grid(column=0, row=5)
        self.ventana1.mainloop()

    def verificar(self):
        cant=0
        if self.seleccion1.get()==1:
            cant+=1
        if self.seleccion2.get()==1:
            cant+=1
        if self.seleccion3.get()==1:
            cant+=1
        self.label1.configure(text="cantidad:"+str(cant))

    def print_name(self):
        print(f'sel1: {self.seleccion1.get()}\n  sel2: {self.seleccion2.get()}\n sel3: {self.seleccion3.get()}')


aplicacion1=Aplicacion()
```
